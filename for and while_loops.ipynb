```python
#Printing the multiplication table 
number  = int(input("enter number: "))

count = 1 #starting number
while count <=10:
    product = number *count
    print(number, "x", count, "=", product)
    count = count +1
```

```python
#to reverse the multiplication table 
number  = int(input("enter number: "))

count = 10#starting number
while count >=1:
    product = number *count
    print(number, "x", count, "=", product)
    count = count -1
```

```python
#for loop to loop over a sequence. A sequence is a collection of items in an order.
number = int(input("Enter the integer: "))


for count in range(1, 101) :
    sum = number + count
    print(number, "+", count, "=", sum)
```
